```python
import pandas as pd

# ======================
# 1 LOAD DATA
# ======================
data_epus = pd.read_csv("clean_epus_tb.csv", low_memory=False)
data_sitb03 = pd.read_csv("clean_sitb03.csv", low_memory=False)
data_sitb06 = pd.read_csv("clean_sitb06.csv", low_memory=False)
icd = pd.read_excel("Kode ICD 10.xlsx")

# ======================
# 2 NORMALISASI NIK
# ======================
def normalize_nik(series):
    return (
        series.astype(str)
        .str.replace(r"\.0$", "", regex=True)
        .str.replace(r"[^0-9]", "", regex=True)
        .str.strip()
    )

data_epus["id"] = normalize_nik(data_epus["id"])
data_sitb03["nomor_identitas_kependudukan_nik"] = normalize_nik(
    data_sitb03.get("nomor_identitas_kependudukan_nik", pd.Series())
)
data_sitb06["nik_nomor_identitas_kependudukan"] = normalize_nik(
    data_sitb06.get("nik_nomor_identitas_kependudukan", pd.Series())
)

# ======================
# 3 KONVERSI TANGGAL
# ======================
data_epus["tanggal_catat"] = pd.to_datetime(data_epus["tanggal_catat"], errors="coerce")

# ======================
# 4 CEK STATUS BERDASARKAN NIK
# ======================
data_epus["status_tb03"] = data_epus["id"].isin(data_sitb03["nomor_identitas_kependudukan_nik"])
data_epus["status_tb06"] = data_epus["id"].isin(data_sitb06["nik_nomor_identitas_kependudukan"])

print(f" Total data ePus: {len(data_epus):,}")
print(f" Ada di TB03: {data_epus['status_tb03'].sum():,}")
print(f" Ada di TB06: {data_epus['status_tb06'].sum():,}")

# ======================
# 5 MERGE DATA BERDASARKAN NIK
# ======================
col_tanggal_03 = next((c for c in data_sitb03.columns if "tanggal_mulai" in c.lower()), None)
col_tanggal_06 = next((c for c in data_sitb06.columns if "tanggal_hasil_diagnosis" in c.lower()), None)

sitb03_slim = data_sitb03[
    [
        c
        for c in [
            "nomor_identitas_kependudukan_nik",
            col_tanggal_03,
            "tipe_diagnosis_tbc",
            "klasifikasi_berdasarkan_lokasi_anatomi",
            "hasil_akhir_pengobatan_hasil",
        ]
        if c in data_sitb03.columns
    ]
].drop_duplicates(subset="nomor_identitas_kependudukan_nik")

sitb06_slim = data_sitb06[
    [
        c
        for c in [
            "nik_nomor_identitas_kependudukan",
            col_tanggal_06,
            "hasil_diagnosis",
            "tipe_diagnosis",
        ]
        if c in data_sitb06.columns
    ]
].drop_duplicates(subset="nik_nomor_identitas_kependudukan")

merged = (
    data_epus
    .merge(sitb03_slim, left_on="id", right_on="nomor_identitas_kependudukan_nik", how="left")
    .merge(sitb06_slim, left_on="id", right_on="nik_nomor_identitas_kependudukan", how="left")
)

# ======================
# 6 ISI NIK KOSONG DARI TB03/TB06 (FALLBACK)
# ======================
merged["id_final"] = merged["id"]
merged.loc[merged["id_final"].isna(), "id_final"] = merged["nomor_identitas_kependudukan_nik"]
merged.loc[merged["id_final"].isna(), "id_final"] = merged["nik_nomor_identitas_kependudukan"]

# ======================
# 7 TAMBAHKAN ICD DAN DESKRIPSI PENYAKIT
# ======================
icd["KD_PENYAKIT"] = icd["KD_PENYAKIT"].astype(str).str.strip().str.upper()
icd["PENYAKIT"] = icd["PENYAKIT"].astype(str).str.strip()

merged["icdx_1_clean"] = merged["icdx_1"].astype(str).str.strip().str.upper()
merged = merged.merge(
    icd[["KD_PENYAKIT", "PENYAKIT"]],
    left_on="icdx_1_clean",
    right_on="KD_PENYAKIT",
    how="left"
)
merged.rename(columns={"KD_PENYAKIT": "kode_icd", "PENYAKIT": "nama_penyakit"}, inplace=True)

# ======================
# 8 FILTER ICD TB (A15A19, Z03.0, Z20.1, Z11.7, Z11.1)
# ======================
tb_icd_codes = [
    "A15","A15.0","A15.1","A15.2","A15.3","A15.4","A15.5","A15.6","A15.7","A15.8","A15.9",
    "A16","A16.0","A16.1","A16.2","A16.3","A16.4","A16.5","A16.7","A16.8","A16.9",
    "A17","A17.0","A17.1","A17.8","A17.9",
    "A18","A18.0","A18.1","A18.2","A18.3","A18.4","A18.5","A18.6","A18.7","A18.8",
    "A19","A19.0","A19.1","A19.2","A19.8","A19.9",
    "Z03.0","Z20.1","Z11.7","Z11.1"
]

filtered_icd = merged[merged["icdx_1_clean"].isin(tb_icd_codes)].copy()

# ======================
# 9 FILTER PUSKESMAS & NIK VALID
# ======================
target_puskesmas = [
    "Puskesmas Bobotsari",
    "Puskesmas Kaligondang",
    "Puskesmas Serayu Larangan",
    "Puskesmas Bojong",
    "Puskesmas Karang Tengah",
    "Puskesmas Karangjambu"
]

filtered = filtered_icd[filtered_icd["puskesmas_names"].isin(target_puskesmas)].copy()
filtered = filtered[filtered["id_final"].notna() & (filtered["id_final"].str.len() > 0)]

# ======================
#  TAMBAHKAN SUMBER DATA (TB03 / TB06 / Keduanya / Tidak ada di SITB)
# ======================
filtered["Sumber_Data"] = filtered.apply(
    lambda x: (
        "Keduanya" if x["status_tb03"] and x["status_tb06"]
        else ("TB03" if x["status_tb03"] else ("TB06" if x["status_tb06"] else "Tidak ada di SITB"))
    ),
    axis=1
)

# Statistik tambahan
only_tb03 = (filtered["Sumber_Data"] == "TB03").sum()
only_tb06 = (filtered["Sumber_Data"] == "TB06").sum()
both = (filtered["Sumber_Data"] == "Keduanya").sum()
none = (filtered["Sumber_Data"] == "Tidak ada di SITB").sum()

print(f" Data ePus ICD TB di 6 puskesmas: {len(filtered):,} baris")
print(f"    Hanya di TB03         : {only_tb03:,}")
print(f"    Hanya di TB06         : {only_tb06:,}")
print(f"    Ada di keduanya       : {both:,}")
print(f"    Tidak ada di SITB     : {none:,}")

# ======================
# 11 LABEL KOLOM SUMBER
# ======================
rename_map = {
    col_tanggal_03: f"{col_tanggal_03} (TB03)" if col_tanggal_03 else None,
    col_tanggal_06: f"{col_tanggal_06} (TB06)" if col_tanggal_06 else None,
    "hasil_diagnosis": "hasil_diagnosis (TB06)",
    "tipe_diagnosis": "tipe_diagnosis (TB06)",
    "tipe_diagnosis_tbc": "tipe_diagnosis_tbc (TB03)",
    "klasifikasi_berdasarkan_lokasi_anatomi": "klasifikasi_berdasarkan_lokasi_anatomi (TB03)",
    "hasil_akhir_pengobatan_hasil": "hasil_akhir_pengobatan_hasil (TB03)"
}
rename_map = {k: v for k, v in rename_map.items() if k in filtered.columns}
filtered.rename(columns=rename_map, inplace=True)

# ======================
# 12 TAMBAHKAN KOLOM DATE, MONTH, YEAR
# ======================
filtered["Date"] = filtered["tanggal_catat"].dt.day
filtered["Month"] = filtered["tanggal_catat"].dt.month
filtered["Year"] = filtered["tanggal_catat"].dt.year

# ======================
# 13 PILIH KOLOM AKHIR
# ======================
final_cols = [
    "nama_pasien",
    "id_final",
    "puskesmas_names",
    "tanggal_catat",
    "Date", "Month", "Year",
    "status_tb03",
    "status_tb06",
    "Sumber_Data",
    rename_map.get(col_tanggal_03, col_tanggal_03),
    rename_map.get(col_tanggal_06, col_tanggal_06),
    "hasil_diagnosis (TB06)",
    "tipe_diagnosis (TB06)",
    "tipe_diagnosis_tbc (TB03)",
    "klasifikasi_berdasarkan_lokasi_anatomi (TB03)",
    "hasil_akhir_pengobatan_hasil (TB03)",
    "icdx_1",
    "kode_icd",
    "nama_penyakit"
]
final_cols = [c for c in final_cols if c in filtered.columns]
final_result = filtered[final_cols].copy()
final_result.rename(columns={"id_final": "id"}, inplace=True)

# ======================
# 14 EXPORT SATU SHEET SAJA
# ======================
output_path = "hasil_epus_tb_6puskesmas_ICDTB_semua_status_with_date.xlsx"
final_result.to_excel(output_path, index=False)

print(f" File final berhasil dibuat: {output_path}")
```
